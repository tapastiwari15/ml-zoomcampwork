```python
rollsroyce = df[df['Make']== "Rolls-Royce"]
rollsroyce.head()
```

```python
rollsroyce = rollsroyce[['Engine HP', 'Engine Cylinders', 'highway MPG']]
```

```python
rollsroyce=rollsroyce.drop_duplicates()
```

```python
X = rollsroyce.values
```

```python
X
```

```python
XT =X.T
XTX = X.T.dot(X)
XTX
```

```python
XTXInv = np.linalg.inv(XTX)
XTXInv
```

```python
XTXInv.sum()
```
